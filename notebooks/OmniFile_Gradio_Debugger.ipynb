## 🔧 الخلية الثانية: واجهة Gradio التفاعلية

هذا الكود مصمم ليعمل كـ **غلاف (Wrapper)** لوظائف مشروعك الأساسية.
يمكنك استبدال دوال المحاكاة بدوال المشروع الحقيقية.

````python
import gradio as gr
import os
import torch
from pathlib import Path

# ============================================================
# إعدادات المشروع
# ============================================================
PROJECT_ROOT = Path("/content/OmniFile_Processor")  # Colab
if not PROJECT_ROOT.exists():
    PROJECT_ROOT = Path(".")  # Local

# إضافة المشروع للمسار
import sys
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

device = "cuda" if torch.cuda.is_available() else "cpu"

# ============================================================
# دالة المعالجة الرئيسية
# ============================================================
def omni_worker(input_file, mode, engine_choice):
    """
    حلقة الوصل بين واجهة المستخدم ومنطق المشروع.
    
    استبدل المحتوى الداخلي بدوال مشروعك الحقيقية:
      from modules.vision.ocr_engine import OCREngine
      from modules.nlp.translator import Translator
      from modules.nlp.text_classifier import TextClassifier
    """
    if input_file is None:
        return "⚠️ الرجاء رفع ملف للمعالجة", ""
    
    file_path = input_file.name
    filename = os.path.basename(file_path)
    ext = filename.rsplit('.', 1)[-1].lower()
    
    # --- تنظيف الذاكرة ---
    if device == "cuda":
        torch.cuda.empty_cache()
    
    try:
        info_lines = [
            f"📂 الملف: {filename}",
            f"📐 الحجم: {os.path.getsize(file_path) / 1024:.1f} KB",
            f"🔧 المحرك: {engine_choice}",
            f"📋 الوضع: {mode}",
            f"💻 الجهاز: {device}",
            f"{'─' * 40}"
        ]
        
        # ===== تصنيف (Classification) =====
        if "تصنيف" in mode:
            # استبدل بـ: from modules.nlp.text_classifier import TextClassifier
            info_lines.append("\n🏷️ نتيجة التصنيف:")
            info_lines.append(f"  النوع المتوقع: مستند {ext}")
            info_lines.append(f"  اللغة: عربية/إنجليزية (يُحدد لاحقاً)")
            info_lines.append(f"  ⏳ ربط مع TextClassifier قيد التطوير...")
        
        # ===== تعرف ضوئي (OCR) =====
        elif "تعرف" in mode or "OCR" in mode:
            # استبدل بـ: from modules.vision.ocr_engine import OCREngine
            info_lines.append(f"\n🔍 نتيجة OCR:")
            info_lines.append(f"  المحرك المختار: {engine_choice}")
            info_lines.append(f"  ⏳ ربط مع OCREngine قيد التطوير...")
            
            # اقتراح: منطق هجين (Hybrid Logic)
            if engine_choice == "Hybrid (EasyOCR + TrOCR)":
                info_lines.append(f"  💡 Hybrid: EasyOCR للكشف + TrOCR للقراءة")
        
        # ===== ترجمة (Translation) =====
        elif "ترجم" in mode:
            # استبدل بـ: from modules.nlp.translator import Translator
            info_lines.append(f"\n🌐 نتيجة الترجمة:")
            info_lines.append(f"  ⏳ ربط مع Translator قيد التطوير...")
        
        result_text = "\n".join(info_lines)
        
        # تنظيف VRAM بعد المعالجة
        if device == "cuda":
            torch.cuda.empty_cache()
        
        return result_text, ""
    
    except Exception as e:
        return f"❌ خطأ أثناء المعالجة:\n{str(e)}", str(type(e).__name__ + ": " + str(e))


# ============================================================
# بناء واجهة غراديو
# ============================================================
with gr.Blocks(
    theme=gr.themes.Soft(),
    title="OmniFile AI Processor",
    css=".container { max-width: 900px; margin: auto; }"
) as demo:
    
    gr.Markdown("# 🧠 OmniFile AI Processor — Interactive Debugger")
    gr.Markdown(
        "نظام متكامل يجمع بين قوة المعالجة اللغوية (NLP) "
        "وتقنيات التعرف الضوئي على الحروف (OCR) مع التركيز على المحتوى العربي."
    )
    
    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### 📂 المدخلات")
            file_input = gr.File(
                label="ارفع الملف (PDF, Image, Docx)",
                file_types=[".pdf", ".png", ".jpg", ".jpeg", ".tiff", ".bmp", ".docx", ".txt"]
            )
            mode_radio = gr.Radio(
                [
                    "تصنيف (Classification)",
                    "تعرف ضوئي (OCR)",
                    "ترجمة (Translation)"
                ],
                label="نوع العملية",
                value="تعرف ضوئي (OCR)"
            )
            engine_radio = gr.Radio(
                [
                    "EasyOCR",
                    "Tesseract",
                    "TrOCR",
                    "PaddleOCR",
                    "Hybrid (EasyOCR + TrOCR)",
                    "Fusion (All Engines)"
                ],
                label="محرك OCR",
                value="EasyOCR"
            )
            submit_btn = gr.Button("🚀 بدء المعالجة", variant="primary", size="lg")
        
        with gr.Column(scale=1):
            gr.Markdown("### 📊 المخرجات")
            output_text = gr.Textbox(
                label="نتيجة المعالجة",
                lines=15,
                show_copy_button=True
            )
            error_text = gr.Textbox(
                label="التتبع (Debug)",
                lines=5,
                visible=True
            )
    
    # ربط الأزرار بالدوال
    submit_btn.click(
        fn=omni_worker,
        inputs=[file_input, mode_radio, engine_radio],
        outputs=[output_text, error_text]
    )
    
    gr.Markdown("---")
    gr.Markdown(
        "### 💡 مقترحات التطوير\n"
        "1. **منطق هجين (Hybrid):** EasyOCR للكشف + TrOCR للقراءة → دقة أعلى في الوثائق الطبية\n"
        "2. **إدارة الذاكرة:** `torch.cuda.empty_cache()` بعد كل عملية معالجة ملف كبير\n"
        "3. **التصحيح اللغوي:** بناء معجم طبي متخصص لمصطلحات الجراحة العظمية\n"
        "4. **فصل المسؤوليات:** فرز الملفات حسب النوع أولاً ثم التحليل العميق فقط للمجهولة"
    )

# ============================================================
# تشغيل الواجهة
    
# share=True ← يعطي رابطاً عاماً (gradio.live) يمكنك فتحه من الجوال
# debug=True ← يظهر الأخطاء (Traceback) مباشرة في المتصفح
demo.launch(share=True, debug=True)
````

## 🔬 مراجعة المشروع ومقترحات التطوير

### 1. إدارة الذاكرة (Memory Management)
- استخدم `device = "cuda" if torch.cuda.is_available() else "cpu` لفحص الـ GPU
- استخدم `torch.cuda.empty_cache()` بعد كل عملية معالجة ملف كبير لتجنب Crash

### 2. تحسين الـ OCR العربي (Hybrid Logic)
- **EasyOCR** → لتحديد مناطق النص (Detection)
- **TrOCR** → لقراءة النص (Recognition)
- هذا يعطي **دقة أعلى** في الوثائق الطبية المعقدة

### 3. التصحيح اللغوي (SymSpell)
- بناء **معجم طبي متخصص** يعتمد على مصطلحات الجراحة العظمية
- القواميس العامة قد تصحح مصطلحات طبية لاتينية/عربية بشكل خاطئ

### 4. هيكلية البيانات (Data Architecture)
- فصل منطق **فرز الملفات** عن منطق **تحليل المحتوى** لتقليل زمن المعالجة
- فرز حسب النوع (Extension) أولاً → ثم التحليل العميق للملفات غير المعروفة فقط

### 5. التتبع (Debugging)
- `debug=True` في `demo.launch()` يُظهر الأخطاء مباشرة في المتصفح
- لا حاجة للعودة للـ Terminal لمعرفة أي دالة فشلت
